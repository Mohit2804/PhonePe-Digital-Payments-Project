**Task 3: Data Quality Checks**


**3.1: Ensure data consistency across state and district levels**


**1. For each state, calculate the total number of transactions, total transaction amount, and total registered users by summing up the values from the district level data.**


**2. Compare the results with the corresponding values at the state level to ensure they match.**


**3. Display any discrepancies found between the district-level and state-level data.**

```python
dist_state_agg = district_txn.groupby(['State', 'Year', 'Quarter']).agg(
    District_Txn = ('Transactions', 'sum'),
    District_Amt = ('Amount (INR)', 'sum'),
    District_Reg_Users = ('Registered Users', 'sum')
).reset_index()

#Comparing the District values with States values
state_level = state_txn[['State', 'Year', 'Quarter', 'Transactions', 'Amount (INR)', 'Registered Users']].copy()
state_level.columns = ['State', 'Year', 'Quarter', 'State_Txn', 'State_Amt', 'State_Reg_Users']

merged = pd.merge(state_level, dist_state_agg, on=['State', 'Year', 'Quarter'])
merged['Txn_diff'] = state_level['State_Txn'] - dist_state_agg['District_Txn']
merged['Amount_diff'] = state_level['State_Amt'] - dist_state_agg['District_Amt']
merged['Reg_u_diff'] = state_level['State_Reg_Users'] - dist_state_agg['District_Reg_Users']

print("\nDiscrepancies:")
discrepancies = merged[
    (merged['Txn_diff'].abs() > 1) |
    (merged['Amount_diff'].abs() > 100) |
    (merged['Reg_u_diff'].abs() > 1)
]

if discrepancies.empty:
    print('No discrepencies - Consistent data')
else:
    print(f'{len(discrepancies)} - discrepancies found:')
    display(discrepancies[['State', 'Year', 'Quarter', 'Txn_diff', 'Amount_diff', 'Reg_u_diff']].head(15).reset_index(drop=True))
```

**4.4: Device brand usage ratio**


**1. Merge the State_DeviceData dataset with the State_Txn and Users dataset.**


**2. Calculate the ratio of users using each device brand to the total number of registered users in each state. Display the results in a tabular format.**


**3. Create a bar chart depicting the device brand usage ratio for each state.**


```python
agg_brand = state_device.groupby(['State', 'Brand'])['Registered Users'].max().reset_index()
state_reg_users = state_txn.groupby('State')['Registered Users'].max().reset_index()
state_reg_users.columns = ['State', 'Total_Registered_Users']
ratio_brand_users = pd.merge(agg_brand, state_reg_users, on='State')

ratio_brand_users['Brand_to_Users_Ratio'] = (ratio_brand_users['Registered Users'] / ratio_brand_users['Total_Registered_Users']).round(4)
ratio_brand_users = ratio_brand_users.sort_values(['State', 'Brand_to_Users_Ratio'], ascending=[True, False])
display(ratio_brand_users.head(20))
```

```python
top_6_brands = (ratio_brand_users.groupby('Brand')['Registered Users'].sum().nlargest(6).index.tolist())
ratio_brand_users = ratio_brand_users[ratio_brand_users['Brand'].isin(top_6_brands)]

top_6_states = state_reg_users.nlargest(6, 'Total_Registered_Users')['State'].tolist()
top_brand = ratio_brand_users[ratio_brand_users['State'].isin(top_6_states)]
pivot_df = top_brand.pivot_table(index='State', columns='Brand', values='Brand_to_Users_Ratio', fill_value=0)

colors = ['#2196F3', '#FF5722', '#4CAF50', '#9C27B0', '#FF9800', '#00BCD4']
bar_chart = pivot_df.plot(kind='bar', figsize=(14, 7), color=colors, edgecolor='white', linewidth=0.9, width=0.8)

for container in bar_chart.containers:
    bar_chart.bar_label(container, fmt='%.2f', fontsize=7, fontweight='bold', padding=2)

plt.title('Device brand usage ratio for each state - Top 6 States', fontsize=13, fontweight='bold')
plt.xlabel('States', fontsize=11, fontweight='bold')
plt.ylabel('Brand_to_Users_Ratio', fontsize=11, fontweight='bold')
plt.xticks(rotation=0)
plt.legend(title='Brand', bbox_to_anchor=(1.01, 1), loc='upper left', fontsize=9)
plt.tight_layout()
plt.show()
```

```python
#or we can use seaborn for above bar chart
plt.figure(figsize=(16, 8))

bar_chart = sns.barplot(data=top_brand, x='State', y='Brand_to_Users_Ratio', hue='Brand',
            palette=['#2196F3', '#FF5722', '#4CAF50', 
                     '#9C27B0', '#FF9800', '#00BCD4'])

for container in bar_chart.containers:
    bar_chart.bar_label(container, fmt='%.2f', fontsize=7, fontweight='bold', padding=2)
    
plt.title('Device brand usage ratio for each state - Top 6 States', fontsize=13, fontweight='bold')
plt.xlabel('States', fontsize=11, fontweight='bold')
plt.ylabel('Brand_to_Users_Ratio', fontsize=11, fontweight='bold')
plt.xticks(rotation=0)
plt.legend(title='Brand', bbox_to_anchor=(1.01, 1), loc='upper left', fontsize=9)
plt.tight_layout()
plt.show()
```
